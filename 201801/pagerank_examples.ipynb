## Modelo Random Walkers

Volvemos a nuestro primer ejemplo de nuestra pequeña representacion de la web con 3 nodos para poder aplicar el modelo de random walkers completo.

<img src="simple_graph.png" width="300px" />

### Representación de Matriz de Links

A partir de ella podemos definir la siguiente matriz de links **M**

```octave
M = [0 1 0; 1/2 0 1; 1/2 0 0]
```

```octave
# M es estocastica? (suma por columna 1)
sum(M,1)
```

### Calculo de primera iteracion version simplificada

- Podemos comenzar con cualquier vector de probabilidades. (via teorema fundamental de las cadenas de Markov). En este caso consideramos todas equiprobables.

```octave
res = M * [1/3; 1/3; 1/3]
```

### Calculo de iteraciones hasta convergencia

```octave
for i =1:100
    res = M * res
end
```

```octave
sum(res,1)
```

## Spider Traps y Teletransportación

Aplicando el modelo redefinido sobre el mismo grafo.

<img src="simple_graph.png" width="300px" />

```octave
M = [0 1 0; 1/2 0 1; 1/2 0 0]
```

### Calculando la matriz G a partir de M

```octave
beta = 0.95 
n = 3
G = beta * M + (1-beta) * [1/n 1/n 1/n; 1/n 1/n 1/n; 1/n 1/n 1/n]
```

```octave
# G es estocastica? (suma por columna 1)
sum(G,1)
```

### Calculo de primera iteracion

```octave
# incializamos de forma equiprobable el vector inicial.
v = [1/n; 1/n; 1/n]
v = G * v
```

### Calculo de iteraciones hasta convergencia

```octave
for i =1:75
    v = G * v
end
```

```octave
sum(v,1)
```
